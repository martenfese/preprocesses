```python
import pandas as pd
import path
import os
```

```python
data_folder = "DataRomandeEnergieVevey"
os.listdir(data_folder)
```

```python
fname = 'equipement.txt'

with open(data_folder+'/'+fname) as f:
    content = f.readlines()

content = [x.strip() for x in content] 
content.append('[END]')

titles = [x for x in content if "[" in x]
titles = titles[2:]
titles

content = content[content.index(titles[0]):]
```

```python
index_start = content.index('[CABLE]')
index_stop = content.index('[CABLE CONCENTRIC NEUTRAL]')

columns = content[index_start+1]
columns = columns.replace("FORMAT_CABLE=", "").split(',')
```

```python
cables = content[index_start+2:index_stop]
cables = [cable.split(',') for cable in cables]
```

```python
df = pd.DataFrame(cables, columns=columns)
df.set_index("ID", inplace=True)
df = df[:-1]

df.to_csv(r'cable.csv', header=True, index=True, sep=' ', mode='a')
df
```
