```python
import pandas as pd
import path
import os
```

```python
data_folder = "DataRomandeEnergieVevey"
```

```python
fname = 'load4exemple.txt'

with open(data_folder+'/'+fname) as f:
    content = f.readlines()

content = [x.strip() for x in content] 
content.append('[END]')


titles = [x for x in content if "[" in x]
titles = titles[2:]


content = content[content.index(titles[0]):]
print(titles)
```

```python
index_start = content.index('[CUSTOMER CLASS]')
index_stop = content.index('[LOADS]')

columns = content[index_start+1]
columns = columns.replace("FORMAT_CUSTOMERCLASS=", "").split(',')

cables_conc = content[index_start+2:index_stop]

cables_conc = [cable.split(',') for cable in cables_conc]

df = pd.DataFrame(cables_conc, columns=columns)
df.set_index("ID", inplace=True)
df = df[:-2]
df
```

```python
index_start = content.index('[LOADS]')
index_stop = content.index('[CUSTOMER LOADS]')

columns = content[index_start+1]
columns = columns.replace("FORMAT_LOADS=", "").split(',')

cables_conc = content[index_start+2:index_stop]

cables_conc = [cable.split(',') for cable in cables_conc]

df = pd.DataFrame(cables_conc, columns=columns)
df.set_index("SectionID", inplace=True)
df = df[:-1]
df
```
