```python
import pandas as pd
import path
import os
```

```python
data_folder = "DataRomandeEnergieVevey"
```

```python
fname = 'network.txt'

with open(data_folder+'/'+fname) as f:
    content = f.readlines()

content = [x.strip() for x in content] 
content.append('[END]')


titles = [x for x in content if "[" in x]
titles = titles[2:]


content = content[content.index(titles[0]):]
print(titles)
```

```python
index_start = content.index('[NODE]')
index_stop = content.index('[HEADNODES]')

columns = content[index_start+1]
columns = columns.replace("FORMAT_NODE=", "").split(',')

cables_conc = content[index_start+2:index_stop]

cables_conc = [cable.split(',') for cable in cables_conc]

df = pd.DataFrame(cables_conc, columns=columns)
df.set_index("NodeID", inplace=True)
df = df[:-1]
df
```
